```python
import os
import json
import pandas as pd
import mysql.connector 
```

```python
def Agg_Trans_Ins(path):
    
    #The os.listdir() method in Python is used to get the list of all files and directories in the specified directory. 
    #If we donâ€™t specify any directory, then a list of files and directories in the current working directory will be returned.

    #Creating dictionary for storing data
    col={'State':[], 'Year':[],'Quarter':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}
    #State
    state_list=os.listdir(path) 
    state_list
    for i in state_list:        
        List=path+i+"/"
        year=os.listdir(List)
    #Year
        for j in year:          
            year_List=List+j+"/"
            yr_list=os.listdir(year_List)
    #File
            for k in yr_list:   # 
                file_List=year_List+k 
                Data=open(file_List,'r')
                D=json.load(Data)    #D 
    #Cleaning the data
                for z in D['data']['transactionData']:
                    col['Transacion_type'].append(z['name'])               
                    col['Transacion_count'].append(z['paymentInstruments'][0]['count'])
                    col['Transacion_amount'].append(z['paymentInstruments'][0]['amount'])
                    col['State'].append(i)
                    col['Year'].append(j)
                    col['Quarter'].append(int(k.strip('.json')))       #1.File name, so to remove . we use strip
    #Succesfully created a dataframe
    df=pd.DataFrame(col)
    df['State']=df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar')
    df['State']=df['State'].str.replace('-'," ")
    df['State'] = df['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli")
    df['State']=df['State'].str.title()
    return df
```

```python
def Agg_User(path):
    #Creating dictionary for storing data
    col={'State':[], 'Year':[],'Quarter':[],'Brand':[], 'Transacion_count':[], 'Percentage':[]}
    #State
    state_list=os.listdir(path) 
    for i in state_list:        
        List=path+i+"/"
        year=os.listdir(List)
    #Year
        for j in year:          
            year_List=List+j+"/"
            yr_list=os.listdir(year_List)
    #File
            for k in yr_list:   # 
                file_List=year_List+k 
                Data=open(file_List,'r')
                D=json.load(Data)    #D 
    #Cleaning the data
                try:
                    for z in D['data']['usersByDevice']:
                        col['Brand'].append(z['brand'])               
                        col['Transacion_count'].append(z['count'])
                        col['Percentage'].append(z['percentage'])
                        col['State'].append(i)
                        col['Year'].append(j)
                        col['Quarter'].append(int(k.strip('.json')))
                except:
                    pass      
    #Succesfully created a dataframe
    df=pd.DataFrame(col)
    df['State']=df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar ')
    df['State']=df['State'].str.replace('-'," ")
    df['State']=df['State'].str.title()
    return df
```

```python
def Map_Trans_Ins(path):
    #Creating dictionary for storing data
    col={'State':[], 'Year':[],'Quarter':[],'District':[], 'Transacion_count':[], 'Transacion_amount':[]}
    #State
    state_list=os.listdir(path) 
    state_list
    for i in state_list:        
        List=path+i+"/"
        year=os.listdir(List)
    #Year
        for j in year:          
            year_List=List+j+"/"
            yr_list=os.listdir(year_List)
    #File
            for k in yr_list:   # 
                file_List=year_List+k 
                Data=open(file_List,'r')
                D=json.load(Data)    #D 
    #Cleaning the data
                for z in D['data']['hoverDataList']:
                    col['District'].append(z['name'])               
                    col['Transacion_count'].append(z['metric'][0]['count'])
                    col['Transacion_amount'].append(z['metric'][0]['amount'])
                    col['State'].append(i)
                    col['Year'].append(j)
                    col['Quarter'].append(int(k.strip('.json')))       #1.File name, so to remove . we use strip
    #Succesfully created a dataframe
    df=pd.DataFrame(col)
    df['State']=df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar ')
    df['State']=df['State'].str.replace('-'," ")
    df['State'] = df['State'].str.replace("Dadra & Nagar Haveli & Daman & Diu", "Dadra and Nagar Haveli")
    df['State']=df['State'].str.title()
    df['District']=df['District'].str.title()
    return df
```

```python
def Map_User(path):
    #Creating dictionary for storing data
    col={'State':[], 'Year':[],'Quarter':[],'District':[],'registeredUsers':[], 'appOpens':[]}
    #State
    state_list=os.listdir(path) 
    state_list
    for i in state_list:        
        List=path+i+"/"
        year=os.listdir(List)
    #Year
        for j in year:          
            year_List=List+j+"/"
            yr_list=os.listdir(year_List)
    #File
            for k in yr_list:   # 
                file_List=year_List+k 
                Data=open(file_List,'r')
                D=json.load(Data)    #D 
    #Cleaning the data
                try:
                    for z in D['data']['hoverData'].items():
                        col['District'].append(z[0])
                        col['registeredUsers'].append(z[1]['registeredUsers'])
                        col['appOpens'].append(z[1]['appOpens'])
                        col['State'].append(i)
                        col['Year'].append(j)
                        col['Quarter'].append(int(k.strip('.json')))
                except:
                    pass       
    #Succesfully created a dataframe
    df=pd.DataFrame(col)
    df['State']=df['State'].str.replace('andaman-&-nicobar-islands','Andaman & Nicobar ')
    df['State']=df['State'].str.replace('-'," ")
    df['State']=df['State'].str.title()
    return df
```

```python
path2="E:/Python/Phonepe/pulse/data/aggregated/user/country/india/state/"
agg_user_list = os.listdir(path2)

columns2 = {"States":[], "Years":[], "Quarter":[], "Brands":[],"Transaction_count":[], "Percentage":[]}

for state in agg_user_list:
    cur_states = path2+state+"/"
    agg_year_list = os.listdir(cur_states)
    
    for year in agg_year_list:
        cur_years = cur_states+year+"/"
        agg_file_list = os.listdir(cur_years)
        
        for file in agg_file_list:
            cur_files = cur_years+file
            data = open(cur_files,"r")
            C = json.load(data)

            try:

                for i in C["data"]["usersByDevice"]:
                    brand = i["brand"]
                    count = i["count"]
                    percentage = i["percentage"]
                    columns2["Brands"].append(brand)
                    columns2["Transaction_count"].append(count)
                    columns2["Percentage"].append(percentage)
                    columns2["States"].append(state)
                    columns2["Years"].append(year)
                    columns2["Quarter"].append(int(file.strip(".json")))
            
            except:
                pass

aggre_user = pd.DataFrame(columns2)

aggre_user["States"] = aggre_user["States"].str.replace("andaman-&-nicobar-islands","Andaman & Nicobar")
aggre_user["States"] = aggre_user["States"].str.replace("-"," ")
aggre_user["States"] = aggre_user["States"].str.title()
```
